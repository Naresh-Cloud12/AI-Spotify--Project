```python
#Data Preprocessing

#What We’ll Do:
#Select useful features (columns) for prediction
#Drop unnecessary columns (like track_id, track_name)
#Encode the genre labels (e.g., pop → 0, rock → 1, etc.)
#Scale the numeric data (so features like tempo and loudness are on the same scale)
#Split the data into training and testing sets

# Reload the dataset to ensure all columns are present
df = pd.read_csv(r"//content/spotify_dataset[1].csv")

# Drop irrelevant columns (they don’t help with prediction)
df = df.drop(columns=[
    'track_id', 'track_name', 'track_artist', 'track_album_id',
    'track_album_name', 'track_album_release_date',
    'playlist_name', 'playlist_id'
])

# Target column
y = df['playlist_genre']

# Features
X = df.drop(columns=['playlist_genre', 'playlist_subgenre'])


#Columns like track_name, artist, or playlist_id don’t help with genre prediction.
#These are just text identifiers.
#playlist_genre is our label (target) → the genre we want to predict.
#X contains all other numerical features to train the model.


#Label Encoding (Convert genres to numbers
from sklearn.preprocessing import LabelEncoder

# Encode playlist_genre (e.g., pop → 0, rock → 1, etc.)
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# To see which number is which genre:
genre_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(genre_mapping)

#Now each genre is mapped to a number.
#Why we use LabelEncoder?
#Because our target column has categories, and we need numerical labels for classification.
```

```python
#Feature Scaling
#Machine learning models work better when all values are in the same range.

#For example:

#Tempo could be 60–180

#Loudness could be –20 to 0
#So we scale them to a common range like 0–1.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#StandardScaler makes all features have mean = 0 and std = 1.
```

```python
#Train-Test Split

#We divide our dataset into:

#Training set (80%) → to train the model

#Testing set (20%) → to test accuracy

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42
)

#We use train_test_split() to avoid overfitting. It ensures the model is tested on unseen data to check real-world accuracy.
```

```python
#Model Building using algorithms like:

#K-Nearest Neighbors

#Decision Tree

#Random Forest

#Accuracy comparison
#------------------
#Model Building & Training
#We’ll now use different machine learning models to train on the training data (X_train, y_train) and predict the genre on the test data.

#We’ll try:

#K-Nearest Neighbors (KNN)

#Decision Tree

#Random Forest
#Then we’ll compare their accuracy and decide the best one.

#1. K-Nearest Neighbors (KNN)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize the model
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_train, y_train)

# Predict
y_pred_knn = knn.predict(X_test)

# Accuracy
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy:", accuracy_knn)

#Explanation:
#KNN predicts a song's genre by looking at its 5 nearest neighbors in feature space.
#If most nearby songs are "pop", it predicts "pop".
#Used for: Simple classification problems, good for smaller datasets
#Not good for: Large datasets or high-dimensional data
```

```python
#2. Decision Tree

from sklearn.tree import DecisionTreeClassifier

# Initialize and train
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

# Predict and accuracy
y_pred_dtree = dtree.predict(X_test)
accuracy_dtree = accuracy_score(y_test, y_pred_dtree)
print("Decision Tree Accuracy:", accuracy_dtree)

#Explanation:
#The decision tree splits the data based on conditions like:

#if tempo > 120 → check valence
#if valence > 0.6 → genre = "pop"

# Good for: Easy interpretation, fast training
# Bad for: Overfitting on small datasets
```

```python
#3. Random Forest

from sklearn.ensemble import RandomForestClassifier

# Initialize and train
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)

# Predict and accuracy
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)

# Explanation:
#Random Forest is like a team of decision trees working together.

#Each tree votes for the genre, and the majority wins.

# Good for: Accuracy, reducing overfitting
# Commonly used in real-world projects
```

```python
#Print All Accuracies Together

print(f"KNN Accuracy         : {accuracy_knn:.2f}")
print(f"Decision Tree Accuracy: {accuracy_dtree:.2f}")
print(f"Random Forest Accuracy: {accuracy_rf:.2f}")
```

```python
#In this project, Random Forest usually gives the best accuracy because it combines multiple trees and handles overfitting well.
```

```python
#Conclusion & Interpretation

#After comparing models, Random Forest Classifier achieved the highest accuracy in predicting Spotify song genres.
#Based on features like tempo, loudness, valence, danceability, the model successfully grouped songs into genres like pop, rock, edm, hip hop, etc.

#Confusion Matrix & Classification Report

#Accuracy tells how many predictions were correct overall.

#But confusion matrix shows which genres were predicted correctly and where mistakes happened.

#Classification report shows:

#Precision: how many predicted pop songs were actually pop?

#Recall: how many actual pop songs did the model find?

#F1-score: balance between precision and recall.

#----
#Confusion Matrix for Random Forest

from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_rf)

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Random Forest")
plt.show()

#Example Output (Heatmap):
#Rows = actual genres

#Columns = predicted genres

#Diagonal = correct predictions
```
